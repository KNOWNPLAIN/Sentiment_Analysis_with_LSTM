### **3.3 EDA**

```python
# Show EDA with sweetviz
my_report = sv.analyze(df)
my_report.show_notebook()
```

```python
# Show correlation between each column
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, fmt='.0%')
```

```python
# Check Outlier
plt.figure(figsize=(10,10))
sns.boxplot(data=df)
```

```python
# Check skewness of each column in dataset and show it in histogram
for column in df.columns:
    if df[column].dtype != object:
        print('Skewness of ' + column + ' : ' + str(df[column].skew()))
        plt.figure(figsize=(10,10))
        sns.distplot(df[column])
        plt.show()
```

#### **3.3.1 Best selling product**

```python
# Show best selling product and show the value in bar chart
df['keyword'].value_counts().head(10).plot(kind='barh', figsize=(10,10))
```

```python
# Make a table for average price in each city
df.groupby('shop.city')['price_int'].mean().sort_values(ascending=False)
```
